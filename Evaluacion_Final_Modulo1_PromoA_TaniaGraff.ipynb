```python
#Creo una clase llamada TiendaOnline que cumpla con los siguientes requisitos:
#La clase TiendaOnline tiene que tener los siguientes atributos: inventario (lista de diccionarios), clientes (diccionario) y ventas_totales (float)
#Comienzo creando la clase y utilizando el método constructor para definir los atributos.

class TiendaOnline:
    #Método constructor
    def __init__ (self, nombre_tienda):
        self.nombre_tienda = nombre_tienda
        self.inventario = [] #Creo una lista vacía de diccionarios para almacenar los datos del inventario...
        #...[{nombre, precio, cantidad}], donde cada producto (con su nombre, precio y cantidad) representará un elemento de la lista.
        self.clientes = {} #Creo un diccionario vacío para almacenar los datos de los clientes {nombre, e-mail, compras}.
        self.ventas_totales = 0 #Creo una variable para llevar un registro de las ventas totales de la tienda.

#Continuo definiendo los métodos o comportamientos de la clase TiendaOnline:
#1. Método agregar_producto
    def agregar_producto(self, nombre, precio, cantidad):
        #Utilizo un bucle for para verificar si el producto ya se encuentra en la lista inventario.
        #El bucle for recorre la lista inventario y valida si el nombre del producto (la primera clave de todos los diccionarios 
        #... almacenados en la lisa inventario) coincide con el nombre que estoy dando.
        for producto in self.inventario:
            #Establezco la condición de que si el producto ya existe en la lista se le sume la nueva cantidad, nombrando la key
            #... de cantidad para sumarle el nuevo valor.
            if producto["nombre"].lower() == nombre.lower():
                producto["cantidad"] += cantidad
                return f"El producto ya estaba en el inventario y se ha actualizado la cantidad: {producto['cantidad']}."
        
        #Si el bucle for no ha encontrado el nombre del producto entre las keys del diccionario, terminará y parará.
        #Añado el nuevo producto a la lista del inventario, como un diccionario.   
        nuevo_producto = {"nombre": nombre, "precio €": precio, "cantidad": cantidad}
        self.inventario.append(nuevo_producto)
        return f"El producto {nuevo_producto} se ha añadido correctamente al inventario."  
    
#2. Método ver_inventario.
    #De esta forma, cuando llamo al método conisgo que me devuelva toda la información del inventario sin necesidad de usar un bucle for.
    def ver_inventario(self):
        return self.inventario
    #Utilizando un bucle for para recorrer la lista me dio varios problemas. Si mantengo el return con productos detrás del bucle,
    #... sólo me devuelve la información del primer diccionario, es decir, el bucle se detiene y no itera por toda la lista de productos.
    #Si dejo el return me devuelve un none al final porque no le he pedido que me devuelva nada. 
    def ver_inventario(self):
        for productos in self.inventario:
            print(productos)
        return self.inventario

#3. Método buscar_producto
    #Vuelvo a necesitar recorrer la lista y acceder a los diccionarios que contiene. Para iterar por ellos, uso el bucle for.
    #Para buscar el producto busco en los diccionarios usando la key de nombre. Si el nombre introducido está en la lista, obtendré 
    #... los detalles del producto.
    def buscar_producto(self, nombre):
        for producto in self.inventario:
            if producto["nombre"].lower() == nombre.lower():
                return f"Los detalles del producto que estás buscando son: {producto}"
        #Añado un mensaje para los productos que no se encuentran dentro del inventario. He probado a mover de posición varias veces
        #... el return porque me daba distintos errores, hasta que lo saqué del bucle y funcionó correctamente.        
        return f"El producto que estás buscando no se encuentra en el inventario."
    
#4. Método actualizar_stock
    #Vuelvo a necesitar recorrer la lista y acceder a los diccionarios que contiene. Para iterar por ellos, uso el bucle for.
    #Para buscar el producto busco en los diccionarios usando la key de nombre. Si el nombre introducido está en la lista,
    #... le pediré al usuario que introduzca una cantidad usando un input.
    #Sumaré la cantidad facilitada por el usario al stock, accediendo desde la Key cantidad.
    def actualizar_stock(self, nombre, cantidad):
        for producto in self.inventario:
            if producto["nombre"].lower() == nombre.lower():
                cantidad = int(input(f"¿Qué cantidad de {producto['nombre']} deseas sumar al stock?"))
                producto["cantidad"] += cantidad
                return f"El stock del producto {producto['nombre']} se ha actualizado y las unidades en stock son de: {producto['cantidad']}."
        #Añado un mensaje para los productos que no se encuentran dentro del inventario fuera del bucle for para que funcione correctamente.    
        return f"El producto que estás buscando no se encuentra en el inventario"
    
#5. Método eliminar_producto
    #Vuelvo a necesitar recorrer la lista y acceder a los diccionarios que contiene. Para iterar por ellos, uso el bucle for.
    #Para buscar el producto busco en los diccionarios usando la key de nombre. Si el nombre introducido está en la lista,
    #... le pediré a la función que elimine el producto de la lista usando remove().
    def eliminar_producto(self, nombre):
        for producto in self.inventario:
            if producto["nombre"].lower() == nombre.lower():
                self.inventario.remove(producto)
                return f"El producto {nombre} se ha eliminado del inventario."
        #Añado un mensaje para los productos que no se encuentran dentro del inventario fuera del bucle for para que funcione correctamente.    
        return f"El producto {nombre} no se encuentra en el inventario."    

#6. Método calcular_valor_inventario
    #Vuelvo a necesitar recorrer la lista y acceder a los diccionarios que contiene. Para iterar por ellos y acceder a las keys de precio y cantidad, 
    #... uso un bucle for.
    #Me doy cuenta de que tengo que crear una variable local para poder almacenar el resultado de multiplicar la cantidad de producto,
    #... por el valor de cada producto.
    def calcular_valor_inventario(self):
        valor_total = 0
        for producto in self.inventario:
            valor_producto = producto["precio €"] * producto["cantidad"]
            valor_total += valor_producto
        return f"El valor total de los productos en inventario es de {valor_total:.2f}€."

#7. Método buscar_producto_regex
    #Vuelvo a necesitar recorrer la lista y acceder a los diccionarios que contiene. Para iterar por ellos y acceder a las keys de precio y cantidad, 
    #... uso un bucle for.
    #Importo import re para poder utilizar las expresiones regulares, creando un patrón que en este caso, me permitirá encontrar todos los
    #... productos en inventario que comiencen por la letra C. (Para que haya más de uno, añadí un nuevo producto en el inventario, camisa flores).
    #El patrón lo comenzé definiendo dentro del método pero al día siguiente, el output no era el esperado (no obtenía ninguno), así que probé a definirlo al llamar al método.
    #Entiendo que de este modo, podré buscar más de un patrón dentro del inventario, porque el patrón de búsqueda puede cambiar cada vez que llame al método.
    #Y que al final tiene más lógica dentro de la creación de un método, ya que todos se construyen de manera que pueda añadir el valor o los valores que
    #... considere al llamarlos. Utilizé search porque para realizar búsquedas me parece el más adecuado.
    def buscar_producto_regex(self, patron):
        import re   
        productos_con_patron = []
        for producto in self.inventario:
            if re.search(patron, producto['nombre']):
                productos_con_patron.append(producto)
        return f"Los productos que coinciden con el patrón {patron} en el inventario son {productos_con_patron}."

#8. Método realizar_compra 
    #Comienzo creando la variable carrito de la compra, una lista vacía donde se añadirán los productos que compre el usuario.
    #Le pregunto al usuario si quiere hacer un pedido y le muestro el inventario para que vea lo que puede comprar.
    #En caso de no querer continuar con la compra, le dejo un mensaje de despedida.
    def realizar_compra(self):
        carrito_compra = []
        costo_total = 0
        self.ver_inventario()
        compra_cliente = input("¿Quieres hacer un pedido? (s/n)")

        if compra_cliente == "n":
            return "Ok, vuelve cuando quieras realizar tu pedido."
    #Mientras el usuario esté comprando, es decir, que la respuesta sea una s de sí, le pediré que introduzca el producto del inventario
    #... que desea comprar y la cantidad. Restaré la cantidad al stock del producto y sumaré el producto a la lista del carrito de la compra.
    #Añado un break para asegurarme de que el while se detiene correctamente y no muestra otros datos del inventario que no deseo mostrar (algo 
    #... que me ocurrió y logré solucionar de esta forma). También le doy la posibilidad al cliente de que vuelva a hacer otro pedido. En caso
    #... de que no quiera hacer un pedido se mostrará el contenido del carrito de la compra.
        else:
            while compra_cliente.lower() == "s":
                compra = input(f"¿Qué productos del inventario quieres comprar?")
                cantidad = int(input("Qué cantidad quieres comprar?"))     
                for producto in self.inventario:
                    if producto["nombre"].lower() == compra.lower() and producto['cantidad']>= cantidad:
                        producto_compra = {"nombre": producto["nombre"], "precio": producto["precio €"],"cantidad": cantidad} #al principio no contaba con ningún
                        #... diccionario pero me di cuenta de que restando la cantidad solicitada por el cliente a producto["cantidad"] y luego lo añadía al carrito
                        #... de la compra, se me añadía con el stock actualizado pero no con la cantidad real solicitada por cliente, así que tuve que crear 
                        #... este otro diccionario, para que en el carrito de la compra se listen correctamente los productos con la información almacenada en pares
                        #... clave/valor.
                        carrito_compra.append(producto_compra)
                        costo_total += producto["precio €"] * cantidad
                        producto["cantidad"] -= cantidad
                        print(f"El producto {producto['nombre']} se ha añadido al carrito de la compra.") 
                        break 
                else:
                    print(f"El producto {producto['nombre']} no se encuentra en el inventario o no hay stock suficiente.")
            
                compra_cliente = input("¿Quieres seguir comprando? (s/n)")  

            return f"El carrito de la compra contiene {carrito_compra} y el total ha pagar es de {costo_total}."

#9. Método procesar_pago
    #Utilizo un bloque try except ValueError para poder calcular el importe a devolver al cliente y advertirle de que la cantidad ingresada
    #... no es correcta en caso de que no utlice valores númericos. Para convertir las comas con las que los usuarios suelen escribir los
    #... céntimos, incluyo en el imput el método replace para cambiar las comas por puntos y que se puedan realizar correctamente las operaciones.
    #Como en input.replace() no me ha permitido incluir un float delate, he convertido a posterirori los valores recogidos en el imput en float 
    #... para poder hacer los cálculos.
    def procesar_pago(self):
        try:
            total = input("Ingresa la cantidad total a pagar:").replace(",", ".")
            pago = input("Ingresa el importe con el que vas a pagar:").replace(",", ".")

            total = float(total)
            pago = float(pago)

            if pago >= total:
                cambio = pago - total
                return f"Tu cambio es de {cambio:.2f}€. Gracias por tu compra."
            elif pago == total:
                return f"Has ingresado la cantidad exacta. Gracias por tu compra."
            elif pago < total:
                cantidad_a_pagar = total - pago
                return f"Te falta por pagar {cantidad_a_pagar:.2}€"
        except ValueError:
            return f"Por favor, ingresa un importe numérico válido."

#10. Método agregar_cliente
    #Busco en el diccionario que contiene los datos de cliente si ya existe el nombre del cliente que estoy buscando. Si el nombre no existe,
    #... lo añado al diccionario, asignándole los pares clave valor correspondientes a los datos que quiero almacenar. Si el cliente ya está
    #... en la base de datos, devuelvo un mensaje para confirmarlo.
    #Sumo el parámetro id_cliente de cara al siguiente ejercicio en el que me pide que muestre la lista de los clientes registrados que, de momento,
    #... almaceno en un diccionario de diccionarios, como se indicó al principio del ejercicio y no en una lista de diccionarios.
    def agregar_cliente(self, id_cliente, nombre, email):
        if id_cliente not in self.clientes:
            self.clientes[id_cliente] = {"nombre":nombre, "email":email}
            return f"La nueva clienta {id_cliente} se ha añadido correctamente a la lista de clientes con su nombre {nombre} y su email {email}."
        else:
            return f"El cliente {id_cliente} ya estaba en la lista de clientes {self.clientes}."

#11. Método ver_clientes. 
    def ver_clientes(self):
        for clientes in self.clientes:
            print(clientes)
        return self.clientes
        
```
